```python
#!pip install textblob pandas csv

from textblob import TextBlob
import pandas as pd
import csv

file_path = 'labelled_newscatcher_dataset.csv'
data = []
with open(file_path, newline='', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile, delimiter=';')
    for row in reader:
        data.append(row)

max_cols = max(len(row) for row in data)
data = [row + [None]*(max_cols - len(row)) for row in data]

df = pd.DataFrame(data)
# Присвоение названий колонкам
df.columns = ['topic', 'link', 'domain', 'published_date', 'title', 'lang']
df = df.drop(index=0)
df.head()
```

```python
# Первичный осмотр данных
print(df.tail())
print(df.info())
```

# определение тональности текстов

```python

def analyze_sentiment(text):
    blob = TextBlob(text)
    sentiment = blob.sentiment
    if sentiment.polarity > 0:
        return sentiment.polarity, 'Позитивный'
    elif sentiment.polarity < 0:
        return sentiment.polarity, 'Негативный'
    else:
        return sentiment.polarity, 'Нейтральный'

df[['polarity', 'sentiment']] = df['title'].apply(lambda x: pd.Series(analyze_sentiment(x)))
df = df.drop(columns=['link', 'domain', 'lang'])

df.head()
```

# фильтрация текста по нужной подстроке ( например компания "вектор") (но так как в примере английские статьи, тоо ище отзывы об Эрике)

```python

# Фильтрация строк, содержащих необходимое упоминание , здесь для примера - "Eric" - т.к. других идей я не нашла
filtered_df = df[df['title'].str.contains("Eric")]

filtered_df
```

```python
filtered_df['published_date'] = pd.to_datetime(filtered_df['published_date'])
import matplotlib.pyplot as plt

sentiment_counts = filtered_df.groupby([filtered_df['published_date'].dt.date, 'sentiment']).size().unstack(fill_value=0)

sentiment_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Изменение отзывов о Charles Barkley по времени')
plt.xlabel('Дата')
plt.ylabel('Количество отзывов')
plt.show()
```

```python
positive_reviews = filtered_df[filtered_df['sentiment'] == 'Позитивный']
if not positive_reviews.empty:
    positive_counts = positive_reviews.groupby(positive_reviews['published_date'].dt.date).size()
    plt.figure(figsize=(10, 6))
    positive_counts.plot(kind='bar', color='green')
    plt.title('Позитивные отзывы о Charles Barkley по времени')
    plt.xlabel('Дата')
    plt.ylabel('Количество отзывов')
    plt.show()
else:
    print("Нет данных для позитивных отзывов")

negative_reviews = filtered_df[filtered_df['sentiment'] == 'Негативный']
if not negative_reviews.empty:
    negative_counts = negative_reviews.groupby(negative_reviews['published_date'].dt.date).size()
    plt.figure(figsize=(10, 6))
    negative_counts.plot(kind='bar', color='red')
    plt.title('Негативные отзывы о Charles Barkley по времени')
    plt.xlabel('Дата')
    plt.ylabel('Количество отзывов')
    plt.show()
else:
    print("Нет данных для негативных отзывов")
```
